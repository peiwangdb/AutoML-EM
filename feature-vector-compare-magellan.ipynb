```python
f_train = []
f_test = []

for i in range(1, 9):
    df_train = pd.read_csv("./02-feature-extract-ori/features/ft_" + str(i) + "_train.csv") 
    df_test = pd.read_csv("./02-feature-extract-ori/features/ft_" + str(i) + "_test.csv") 
    f_train.append(df_train)
    f_test.append(df_test)
    print(len(df_train), len(df_test), len(df_train.columns.to_list()) - 2)
```

```python
# compare the test score

for i in range(0, 8):
    print('round:', i)
    
    df_train_m = m_train[i]
    df_test_m = m_test[i]
    
    print(len(df_train_m), len(df_test_m), len(df_train_m.columns.to_list()) - 2)
    
    X_train_m = np.array(df_train_m.drop(['label'], axis=1).values.tolist())
    y_train_m = np.array(df_train_m['label'].values.tolist())
    X_test_m = np.array(df_test_m.drop(['label'], axis=1).values.tolist())
    y_test_m = np.array(df_test_m['label'].values.tolist())
    
    automl = autosklearn.classification.AutoSklearnClassifier(
            include_estimators=['random_forest'], 
            time_left_for_this_task=3600, 
            initial_configurations_via_metalearning=0, 
            seed = 1,
            ensemble_nbest=1, 
            ensemble_size=1, 
            resampling_strategy='holdout',
            resampling_strategy_arguments={'train_size':0.75, 'random_state':42}
        )
    automl.fit(X_train_m, y_train_m, metric=autosklearn.metrics.f1)
    print("Magellan: training fscore:", automl.score(X_train_m, y_train_m),  
              ", test fscore: ", automl.score(X_test_m, y_test_m), '\n')
    print(automl.show_models())
    print(automl.sprint_statistics())
    
#     df_train_f = f_train[i]
#     df_test_f = f_test[i]
    
#     print(len(df_train_f), len(df_test_f), len(df_train_f.columns.to_list()) - 2)
    
#     X_train_f = np.array(df_train_f.drop(['label'], axis=1).values.tolist())
#     y_train_f = np.array(df_train_f['label'].values.tolist())
#     X_test_f = np.array(df_test_f.drop(['label'], axis=1).values.tolist())
#     y_test_f = np.array(df_test_f['label'].values.tolist())
    
#     automl = autosklearn.classification.AutoSklearnClassifier(
#             include_estimators=['random_forest'], 
#             time_left_for_this_task=3600, 
#             initial_configurations_via_metalearning=0, 
#             seed = 1,
#             ensemble_nbest=1, 
#             ensemble_size=1, 
#             resampling_strategy='holdout',
#             resampling_strategy_arguments={'train_size':0.75, 'random_state':42}
#         )
#     automl.fit(X_train_f, y_train_f, metric=autosklearn.metrics.f1)
#     print("AutoML-EM: training fscore:", automl.score(X_train_f, y_train_f),  
#               ", test fscore: ", automl.score(X_test_f, y_test_f), '\n')
#     print(automl.show_models())
#     print(automl.sprint_statistics())
    
    
    print('\n\n\n')
```
